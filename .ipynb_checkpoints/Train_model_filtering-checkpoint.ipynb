```python
with open('Slang_data\en.txt','r',encoding='utf-8') as cens_lang:
    cens_word = set()
    for index,key in enumerate(cens_lang):
        cens_word.add(key.strip())
cens_word
```

```python
data_twitter['vulgar'] = None
```

```python
data_twitter
```

```python
import re

# Initialize column with default value 0 (not vulgar)
data_twitter["vulgar"] = 0  

for idx, key in enumerate(data_twitter["tweet"]):  
    if pd.isna(key):  # Skip if NaN (missing tweet)
        continue

    text_list = key.split()
    for word in text_list:
        word_search = re.search(r"\b[A-Za-z]+\b", word)
        if word_search:
            extr_word = word_search.group(0)
            if extr_word.lower() in cens_word:
                data_twitter.at[idx, "vulgar"] = 1  # Update only the relevant row
                break  # Stop checking after finding the first vulgar word

# Save the updated dataset (optional)
data_twitter.to_csv("traintest/train_labeled.csv", index=False)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of 0 and 1
vulgar_counts = data_twitter["vulgar"].value_counts()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(vulgar_counts, labels=["Not Vulgar", "Vulgar"], autopct="%1.1f%%", colors=["green", "red"], startangle=90)
plt.title("Proportion of Vulgar vs Non-Vulgar Tweets")
plt.show()
```
